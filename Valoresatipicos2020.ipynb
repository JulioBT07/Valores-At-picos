```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Agregar el archivo: Gastos y costos 20-23.xlsx

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2020', header = 5)
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
data1 =data.fillna(method="bfill")
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data1 = data1.drop(['POLIZA'], axis=1)


data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Dataframe1.csv")
```

##Aplicar métodos (Desviación Estándar y Rango Intercuartílico) para eliminar Outliers, comprobar con diagramas de caja de cada dataframe

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
print(data2.shape)
```

```python
#Convertir DataFrame a CSV
data2.to_csv("Dataframe2.csv")
```

```python
data2
```

```python
columnas_a_eliminar = ['FECHA', 'FOLIO', 'UUID', 'RFC', 'PROVEEDOR', 'TIPO GASTO', 'GASTO', 'DESCRIPCION', 'TIPO','STATUS', 'Columna1']
data3 = data2.drop(columnas_a_eliminar, axis=1)
```

```python
#Convertir DataFrame a CSV
data3.to_csv("Dataframe3.csv")
```

```python
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Comprobamos con diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

Observamos que los datos se encuentran dentro de los limites de tres desviaciones estandar
