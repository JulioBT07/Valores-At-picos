```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##Agregar el archivo: Gastos y costos 20-23.xlsx

```python
#Carga desde un archivo .csv sin indice
data= pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2023', header = 3)
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
data1 =data.fillna(method="bfill")
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data1['MP'] =data1['MP'].fillna(method="ffill")
data1['FP'] =data1['FP'].fillna(method="ffill")
data1['Otros'] =data1['Otros'].fillna(method="ffill")
data1['Poliza'] =data1['Poliza'].fillna(method="ffill")
data1 = data1.drop(['Otros'], axis=1)
data1
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data1.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

```python
#Convertir DataFrame a CSV
data1.to_csv("Dataframe1.csv")
```

##Aplicar métodos (Desviación Estándar y Rango Intercuartílico) para eliminar Outliers, comprobar con diagramas de caja de cada dataframe

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data1
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data2= data1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
print(data2.shape)
```

```python
#Convertir DataFrame a CSV
data2.to_csv("Dataframe2.csv")
```

```python
columnas_a_eliminar = ['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO', 'Descripción', 'MP', 'Tipo','Status', 'Poliza']
data3 = data2.drop(columnas_a_eliminar, axis=1)
```

```python
#Convertir DataFrame a CSV
data3.to_csv("Dataframe3.csv")
```

```python
data3
```

```python
#Corroboramos valores nulos del dataframe4
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
#Comprobamos con diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data3.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

Observamos que los datos se encuentran dentro de los limites de tres desviaciones estandar

Agregamos de nuevo un indice que en este caso es la variable folio.

```python
folio=data['Folio']
data_clean = pd.concat([folio, data_clean], axis=1)
data_clean
```

**PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME**

```python
#Método aplicando Cuartiles. Encuentro cuartiles 0.25 y 0.75
data4 = data2.copy()
columnas_a_eliminar = ['Fecha', 'Folio', 'UUID', 'RFC', 'Proveedor', 'TIPO GASTO', 'Descripción', 'MP', 'Tipo','Status', 'Poliza'] #Eliminamos columnas numericas
data4 = data4.drop(columnas_a_eliminar, axis=1)
y=data4

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Convertir DataFrame a CSV
data4.to_csv("Dataframe4.csv")
```

```python
#Obtenemos datos limpios del Dataframe
data4_iqr= data4[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
data4_iqr
```

```python
#Convertir DataFrame a CSV
data4_iqr.to_csv("Dataframe4_iqr.csv")
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data4_iqr.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data4_iqr=data4_iqr.copy()
data4_iqr=data4_iqr.fillna(round(data4_iqr.mean(),1))
data4_iqr
```

```python
#Comprobamos con diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data4_iqr.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

Observamos que los datos se encuentran dentro de los limites que obtuvimos dentro de los cuantiles de 0.25 y 0.75

```python
Datos_limpios = data4_iqr
Datos_limpios
```

Agregamos de nuevo un indice que en este caso es la variable folio.

```python
folio=data['Folio']
Datos_limpios = pd.concat([folio, data4_iqr], axis=1)
Datos_limpios
```
